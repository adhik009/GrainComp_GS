## Comparing prediction accuracy using % Dry Basis values to values per grain
The original NIR values were obtained in Dry Basis  which was transformed to values per grain using the following formula:
 Pheno_perGrain = Pheno_DB * per grain weight ( using TGW/1000 )
 Values obtained in per gram for all traits except 'Gross Energy' which is in Calories.

For each methods the blups (adjusted for random effect of replication within environment) for each environment were used for prediction in BGLR or BMTME R packages.

### Single Trait Single Environment Prediction
Script used: 'STSE_BGLR_CV5.qsub'
Analysis done: Five fold cross validation setting seed from 123 to 132 (a total of 10 iterations and five folds per run).

```r
df <- read.csv('results/STSE/SAP_CV5_STSE_MethodsCompiled.csv')
df$Env <- as.factor(df$Env)

df <- df[df$Trait != 'Amylose',]
head(df)
png('results/STSE/SAP_STSE_DBvsPerGrain_accuracy.png', units='in',width=7,height=7, res=500)
ggplot(df,aes(x=Env,y=r,fill=Method)) + geom_boxplot() + facet_wrap(Trait~., scale='free') + theme_bw() + labs(x='Year',y='Predictive ability (r)')
dev.off()
```

### Bayesian Multi-output Regression Stacking (BMORS)
This method uses multi-target (multivariate) regression to implement multi-trait multi-environment prediction.
Implemented using BMORS() and BMORS_Env() functions in BMTME package. Similar to BMTME but much more efficient and is machine learning approach.

#### BMORS using cross validation to get accuracy
script: BMORS.qsub
Analysis: CV5, seed = 123, ETA using BRR, covariate models = BayesB

```r
df <- read.csv('results/BMORS/SAP_CV5_BMORS_MethodsCompiled.csv',header=T)
df$Environment <- as.factor(df$Environment)
head(df)
ggplot(df,aes(x=Environment,y=Pearson,fill=Method)) + geom_boxplot() + facet_wrap(Trait~., scale='free') + theme_bw() + labs(x='Year',y='Prediction Accuracy (r)')
```

#### BMORS using BMORS_Env to prediction entire environment
script: BMORS.qsub
Analysis: Predict whole environment, the model uses all other environment to predict one whole environment

```r
df <- read.csv('results/BMORS/SAP_CV-ENV_BMORS_MethodsCompiled.csv',header=T)
df$Environment <- as.factor(df$Environment)
head(df)
ggplot(df,aes(x=Environment,y=Pearson,fill=Method)) + geom_bar(stat='identity',position=position_dodge()) + facet_wrap(Trait~., scale='free') + theme_bw() + labs(x='Year',y='Prediction Accuracy (r)')
```
